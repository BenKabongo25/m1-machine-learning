```python
import numpy as np
import matplotlib.pyplot as plt
from tme3 import *
from mltools import plot_data, plot_frontiere, make_grid, gen_arti

check_fonctions()
```

## Expérimentations

### Régression linéaire

#### Cas de base

```python
# données aléatoires
X, y = gen_arti(nbex=1000, data_type=0, epsilon=0.02)
plot_data(X, y)
```

```python
eps=1e-4
max_iter=1_000
w_lin, all_w_lin, loss_lin = descente_gradient(X, y, mse, mse_grad, eps, max_iter)
```

```python
w_lin
```

```python
# frontière de décision
plot_frontiere(X, lambda x : np.sign(x.dot(w_lin)), step=100)
plot_data(X, y)
```

```python
# frontière de décision
plot_frontiere(X, lambda x : np.sign(x.dot(w_lin)), step=1000)
plot_data(X, y)
```

```python
# fonction de cout
grid, x_grid, y_grid = make_grid(xmin=-2, xmax=2, ymin=-2, ymax=2, step=100)
plt.contourf(x_grid, y_grid, 
             np.array([mse(w, X, y).mean() for w in grid]).reshape(x_grid.shape),
             levels=20)
```

```python
# évolution du coût en fonction des itérations
plt.plot(loss_lin)
```

#### Impact du pas d'apprentissage

```python
# étude du pas d'apprentissage
for eps in [1e-4, 1e-3, 1e-2, 1e-1]:
    w_lin, all_w_lin, loss_lin = descente_gradient(X, y, mse, mse_grad, eps, max_iter)
    plt.plot(loss_lin, label=f'eps={eps}')
plt.legend()
```

```python
# étude du pas d'apprentissage
max_iter=10_000
for eps in [1e-4, 1e-3, 1e-2, 1e-1]:
    w_lin, all_w_lin, loss_lin = descente_gradient(X, y, mse, mse_grad, eps, max_iter)
    plt.plot(loss_lin, label=f'eps={eps}')
plt.legend()
```

#### Données bruités

```python
# données bruitées
max_iter=1_000
for epsilon in np.arange(0, 0.1, 0.01):
    X, y = gen_arti(nbex=1250, data_type=0, epsilon=epsilon)
    w_lin, all_w_lin, loss_lin = descente_gradient(X, y, mse, mse_grad, eps, max_iter)
    plt.plot(loss_lin, label=f'bruit={epsilon}')
plt.legend()
```

#### USPS

```python
import numpy as np

def load_usps(fn):
    with open(fn,"r") as f:
        f.readline()
        data = [[float(x) for x in l.split()] for l in f if len(l.split())>2]
    tmp=np.array(data)
    return tmp[:,1:],tmp[:,0].astype(int)

def get_usps(l,datax,datay):
    if type(l)!=list:
        resx = datax[datay==l,:]
        resy = datay[datay==l]
        return resx,resy
    tmp =   list(zip(*[get_usps(i,datax,datay) for i in l]))
    tmpx,tmpy = np.vstack(tmp[0]),np.hstack(tmp[1])
    return tmpx,tmpy

def show_usps(data):
    plt.imshow(data.reshape((16,16)),interpolation="nearest")
```

```python
uspsdatatrain = "../TME04/data/USPS_train.txt"
uspsdatatest = "../TME04/data/USPS_test.txt"
alltrainx,alltrainy = load_usps(uspsdatatrain)
alltestx,alltesty = load_usps(uspsdatatest)
neg = 6
pos = 9
X_train, y_train = get_usps([neg,pos],alltrainx,alltrainy)
X_test, y_test = get_usps([neg,pos],alltestx,alltesty)
y_train = np.where(y_train == neg, -1, 1)
y_test = np.where(y_test == neg, -1, 1)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
svc = SVC()
svc.fit(X_train, y_train)
print('Train accuracy : ', accuracy_score(svc.predict(X_train), y_train))
print('Test accuracy : ', accuracy_score(svc.predict(X_test), y_test))
```

```python
for vector in svc.support_vectors_:
    show_usps(vector)
```

```python
svc = SVC(kernel='poly')
svc.fit(X_train, y_train)
print('Train accuracy : ', accuracy_score(svc.predict(X_train), y_train))
print('Test accuracy : ', accuracy_score(svc.predict(X_test), y_test))
```

```python
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
print('Train accuracy : ', accuracy_score(svc.predict(X_train), y_train))
print('Test accuracy : ', accuracy_score(svc.predict(X_test), y_test))
```

#### Noyau et vecteurs supports

```python
X, y = gen_arti(nbex=1250, data_type=0, epsilon=.0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
# noyau polynomial
degrees = range(2, 17, 2)
train_accs = []
test_accs = []
n_vectors = []

for degree in degrees:
    p = SVC(degree=degree, kernel='poly')
    p.fit(X_train, y_train)
    train_acc = accuracy_score(p.predict(X_train), y_train)
    test_acc = accuracy_score(p.predict(X_test), y_test)
    n_vector = len(p.support_vectors_)
    train_accs.append(train_acc)
    test_accs.append(test_acc)
    n_vectors.append(n_vector)
    print('Poly degree : ', degree)
    print('\tTrain accuracy : ', train_acc)
    print('\tTrain accuracy : ', test_acc)
    print('\tN vectors : ', n_vector)
    
```

```python
plt.plot(degrees, train_accs, label='Train accuracy')
plt.plot(degrees, test_accs, label='Test accuracy')
plt.xlabel('degree')
plt.ylabel('accuracy')
plt.legend()
```

```python
plt.plot(degrees, n_vectors, label='N vectors')
plt.xlabel('degree')
plt.ylabel('n vectors')
plt.legend()
```

```python
# noyau rbf
gammas = [1e-3, 1e-2, 1e-1, 1, 1e+1, 1e+2, 1e+3]
train_accs = []
test_accs = []
n_vectors = []

for gamma in gammas:
    p = SVC(gamma=gamma, kernel='rbf')
    p.fit(X_train, y_train)
    train_acc = accuracy_score(p.predict(X_train), y_train)
    test_acc = accuracy_score(p.predict(X_test), y_test)
    n_vector = len(p.support_vectors_)
    train_accs.append(train_acc)
    test_accs.append(test_acc)
    n_vectors.append(n_vector)
    print('Gamma : ', gamma)
    print('\tTrain accuracy : ', train_acc)
    print('\tTrain accuracy : ', test_acc)
    print('\tN vectors : ', n_vector)
    
```

```python
plt.plot(gammas, train_accs, label='Train accuracy')
plt.plot(gammas, test_accs, label='Test accuracy')
plt.xlabel('gamma')
plt.ylabel('accuracy')
#plt.xticks(gammas, rotation=90)
plt.legend()
```

```python
plt.plot(gammas, n_vectors, label='N vectors')
plt.xlabel('gamma')
plt.ylabel('n vectors')
plt.legend()
```

#### Grid search

```python
X, y = gen_arti(nbex=1250, data_type=0, epsilon=.0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

param_grid = [
    {'kernel': ['linear'], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['poly'], 'degree': [2, 3, 4], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['rbf'], 'gamma': [0.1, 1, 10, 100], 'C': [0.1, 1, 10, 100]}
]

svm = SVC()

grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best params : ", grid_search.best_params_)

best_svm = grid_search.best_estimator_

#train_scores = grid_search.cv_results_['mean_train_score']
test_scores = grid_search.cv_results_['mean_test_score']
params = grid_search.cv_results_['params']

print(*params, sep='\n')
plt.plot(range(len(test_scores)), test_scores)
plt.xticks(range(len(test_scores)))
plt.show()
```
